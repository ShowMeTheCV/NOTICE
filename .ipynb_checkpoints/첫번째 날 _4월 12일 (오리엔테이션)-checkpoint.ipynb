## 디지털 이미지 프로세싱

## 1. 컴퓨터 비전이란?   

- 정의 : 컴퓨터를 이용하여 정지 영상(사진) 혹은 동영상으로부터 의미있는 정보를 추출하는 방법을 연구하는 학문          
- 즉, 사람의 눈이 하는 작업을 컴퓨터로 할 수 있게끔 하는 작업입니다. (컴퓨터에게 눈을 달아주는 작업이라고 할 수 있겠죠!)    

그럼 어떻게 컴퓨터에게 눈을 달아줄 수 있을까요?   
전통적으로, 컴퓨터 비전에서는

> 1. 영상으로부터 유용한 정보를 추출하고      
> 2. 이를 조합하여 결과를 유추합니다.  

아래의 사진의 경우를 생각해볼까요?

![사과한개](./PostingPic/apple.jpeg)

- 컴퓨터는 이 사진이 사과라는 것을 어떻게 알 수 있을까요?              
: 흰 배경을 제외한 중앙에 빨간색 값이 많고, 둥근 형태이며, 꼭지가 있으면 -> 사과!


- 그럼 초록색 사과는?       
- 그럼 빨갛고 둥근 토마토는?

![사과들](./PostingPic/apples.jpg)

- 이렇게 사과가 우르르르르르 있는 경우는?

__컴퓨터 비전에서 주로 활용하는 영상정보는 밝기, 색상, 모양, 텍스쳐 등이 있으며, 이러한 정보와 머신러닝을 함께 사용하여 사물을 인지할 수 있습니다.__   

그러나, 위에서 살펴 본 것처럼        

1. 배경과 객체를 어떻게 구분해야 하는지       
2. 빨간색임을 알기 위해 어떤 수식을 사용해야 하는지         
3. 모양이 둥근지 네모난지          
4. 사과와 토마토를 어떻게 구분하는지     

알기 위해서는 어떤 알고리즘이 적합한지 결정하기가 쉽지 않습니다. 심지어 사진의 컨디션에 의해서도 영향을 받기 마련이죠.    

1. 날씨나 시간대에 따른 조명 변화       
2. 카메라 시점의 변화            
3. 잡음(블러, 노이즈)           

의 영향으로 이미지나 영상이 항상 확실한 정보를 제공하리라고 확신할 수 없습니다. 

> 물론, 처음부터 컴퓨터비전이 이 모든것을 한번에 인식할 수는 없었을 것입니다.            
> 컴퓨터비전이 어떻게 발전해왔는지, 한 번 살펴볼까요?

## 2. 컴퓨터 비전의 발전사

####  1. 1960년대          
컴퓨터가 보급되기 시작한 1960년대부터, 컴퓨터 비전이 발전하기 시작합니다.(1960' 이전에는 필름을 오려서 붙이는 합성을 했다구요!)      

- 1960년대, 미국에서 인공위성으로부터 전송된 달 표면 사진의 잡음을 제거하는 작업을 수행한 것이 디지털 영상 처리의 시초랍니다.    
- 1966년에는 MIT 대학에서 'The Summer Vision Project' 라는 이름으로 카메라가 보는 뷰를 컴퓨터가 인식하는 프로젝트를 시작했구요.  


#### 2. 1970~1980년대     
영상에서 객체의 윤곽, 엣지를 추출하거나 코너 점 검출, 모양 정보 분석, 텍스쳐 분석하기(기본적인 영상 특징 분석)      
3차원 구조 분석, 움직임 정보 추출, 얼굴 검출 및 인식 등       


####  3. 2000년대          
실시간 얼굴 검출, 크기 불변 특징점 검출 및 매칭 기술들이 개발되며 컴퓨터 비전 기술이 현실에 스며들기 시작했답니다.      


#### 4. 2010년대  
__딥러닝의 출현으로 컴퓨터 비전의 가능성이 확대되고 있습니다!__   지금 우리가 배우는 이런 다양한 태스크처럼 말이죠!     

## 3. 컴퓨터 비전, 어디에 쓰이는가?          

컴퓨터 비전이 활용되는 분야는 무궁무진합니다.

1. 스마트폰 카메라의 HDR 사진 촬영, 인스타그램의 필터 기능 등 엔터테인먼트적인 목적       
2. 제조 상품의 검사, 측정, 불량 판정 등의 생산 보조적 목적(머신 비전이라고 합니다)        
3. 길거리에 설치된 CCTV로 범죄자 얼굴을 인식하는 기술        
4. 자율주행 기술 등....!

## 4. 우리가 분석할 영상, 어떻게 구성되어 있는지 한 번 살펴봅시다.       

__여기서 잠깐! CV에서__    

영상 : == 이미지      
우리가 흔히 생각하는 그 '영상' : == 동영상   으로 표현한답니다.

### 4-1. 영상이 인식되는 순서       

![영상획득과정](./PostingPic/process.jpg)

- 영상은 '빛을 전기적 신호로 변환하는 다이오드 센서(포트다이오드)' 를 통해, 빛 반사에 따른 2차원 명암 비율을 얻는 것으로 인지됩니다.     
- 포트 다이오드에서 생성된 전기 신호는 아날로그-디지털 변환기를 통해 디지털 신호로 바뀌고,         
- 이 신호는 카메라의 ISP(Image Signal Processor)로 전달되어 기초 처리를 수행한 후 2차원의 디지털 영상(이미지)를 생성합니다.    
- 이 영상(이미지)은 필요에 따라 JPT, TIFF 등의 영상 파일 형식으로 변환되어 저장됩니다.

### 4-2. 영상을 표현하는 방법

![영상좌표계](./PostingPic/upAndDown.jpg)

- 영상은 우리가 흔히 알고 있듯이, x와 y로 구성된 2차원 좌표계로 표현됩니다.     
- 이걸 0-기반 표현방식이라고 합니다.(zero-based)     


- 단, x는 우리의 인식대로 (작음)왼쪽 --------------> 오른쪽(큼) 이지만, (0:w-1)      
- y는 (작음) 위--------------------------------> 아래(큼) 랍니다. (0:h-1)        

__X, Y의 2차원 값을 다룰 때 이제 자연스레 '행렬'이 생각나지 않나요? 맞습니다!__       

- 영상을 2차원 행렬로 표현하기도 합니다. 
- 몇 몇 영상처리 알고리즘들은, 행렬 이론을 이용하여 컴퓨터 비전 문제를 해결하기도 합니다.     
- 이 경우, 다음과 같이 나타냅니다.      


![행렬로](./PostingPic/matrix.jpg)

> OpenCV의 크기 표현에서 주의할 점 :      

가로X세로($w * h$)의 표현 방식과, 행렬 표현 방식은 반대입니다. 표기에 주의합시다. :)    

![행렬](./PostingPic/행렬.png)

### 4-3. 그레이스케일 영상과 컬러 영상

- 그레이스케일 영상 : 흑백(x) 오직 밝기 정보만으로 구성된 영상, 회색조 영상이라고도 합니다.       
- 트루컬러 영상 : 256개 이상의 색상으로 구성된 컬러 영상(256개 이하도 있어서 구분했지만, 요즘에는 굳이 구분하지 않는답니다.)     

![scale](./PostingPic/scale.jpg)

> 책에 있는 이미지를 인터넷에서 찾으니 그레이스케일로만 주더라고요.. (착한 어른이는 컬러로 보입니다.)

- 그레이스케일 영상은 몇 개의 단계로 나타날까요?        
- 놀랍게도 그레이스케일도 동일하게 256개의 단계를 갖습니다. 단, (0)검은색 ------ 흰색(255) 의 값을 갖는 것 뿐이죠.          


- 트루컬러는 여러분들이 아시는대로 RGB의 3채널 색상을 0에서~255까지 조합하여 표현하지만, __OpenCV에서는 RGB가 아니라 BGR__ 입니다!

![rgb](./PostingPic/rgb.jpg)

## 5. OpenCV 개요          

### 5-1. OpenCV 란?         

- Open Source Computer Vision Library
- 오픈소스로 개발되고 있는 컴퓨터 비전 및 머신러닝 라이브러리입니다.          
- 2500개가 넘는 컴퓨터 비전 알고리즘과 머신러닝 알고리즘을 포함하고 있습니다.            

- OpenCV로 할 수 있는 것      

1. 영상 파일 입출력     
2. 영상의 화질 향상        
3. 객체 검출과 인식, 추적       
4. 3차원 비전 해결 문제          
+ 5. 머신러닝 알고리즘(K최근접 이웃, 서포트 벡터 머신 등)         

[궁금하시다면 공식 사이트로 :](https://opencv.org/)                   


### 5-2. 꽤 유용하게 쓰인답니다              

- BSD 라이선스를 따르고 있어, 연구용 및 상업 용도로 자유롭게 사용 가능합니다.          

구글, 야후, 마이크로소픝, 인텔, IBM, 소니, 혼다, 토요타에서도 OpenCV를 활용하여 서비스들을 만들어내고 있죠!

### 5-5. 우리가 OpenCV로 진행할 것         

##### 1. 전처리 알고리즘       
- 이미지는 매우 많은 데이터로 구성되어 있기 때문에, 불필요한 데이터를 줄이고 유의미한 데이터를 정제해야 합니다.     
- 이 때, __검출(목적에 맞는 결과 도출)에 방해가 되는 요소들을 최소화하는 것을 전처리라고 합니다.__       

- 데이터 폭 줄이는 알고리즘(노말라이제이션)     
- 특정 지점을 기준으로 데이터를 가공하는 알고리즘(바인딩 박스)   
- 데이터를 검출 쉬운 상태로 변형하는 알고리즘(데이터 반전 등)        

##### 2. 노이즈 및 디노이즈     
- 노이즈 : 알고리즘 상에서 원하지 않는 데이터. 다른 데이터를 간섭하거나, 의도하지 않은 왜곡      
- 디노이즈 : 노이즈를 제거하여 검출/분석의 정확성과 정밀도를 높이는 작업      

- 이미지 복원 : 훼손된 영역이나 객체를 조작하거나 변형해서 본래 형태로 복구하는 작업 
- 블러링    
- 샤프닝    
- 평균값 대체 등      

![솔트앤페퍼](./PostingPic/saltandpepper.png)        

##### 3. 특징 및 유사성 검출     

- 특징점 검출 : 이미지 내의 주요 특징점을 검출하는 방법. 특징의 위치를 알려주거나, 특징을 더 부각시킵니다.       
- 픽셀의 색 강도, 연속성, 변화량, 의존성, 유사성, 임계점 등을 사용해 파악하며, 특징 검출로 다양한 패턴의 객체를 검출할 수 있습니다.     


- 유사성 검출 : 이미지 내의 주요한 유사 영역을 검출하는 방법. 비슷한 영역을 알려주거나, 영역을 부각시킵니다.         
- 유사모델(positive image)와 전혀 다른 이미지(negative image)로 훈련된 모델을 사용하거나,       
- 이미지의 크기를 변형시켜 이미지에서 일치하는 영역이 큰 영역을 찾습니다.
