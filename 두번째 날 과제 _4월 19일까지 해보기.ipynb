### 1. 이미지의 크기 속성         

- 이미지 크기란, 필드나 변수에 할당될 이미지의 너비 * 이미지의 높이를 의미한다. 이 __이미지는 행렬(matrix)형태로 구성되어 있다.__   
- 많은 알고리즘에서 이미지 크기를 변경하는 메서드인 Image Pyramide나 Resize를 전처리 과정에서 적용한다.   
- __단순히 연산할 이미지의 크기를 바꾸는 과정도 딥러닝의 정확도와 연산량에 큰 영향을 미친다.__      


- openCV에서 변수나 필드에서 __이미 설정된 이미지의 크기로 다른 이미지를 불러올 경우 오류가 나곤 하는데, 이는 이미지 크기 속성 값이 "액자 역할" 을 하기 때문이다.    
- 불러온 이미지의 크기와 액자의 크기가 다르다면 오류가 날 수 밖에 없다 ㅠㅠㅠ..

### 2. 정밀도    

- __정밀도__ 란 : __이미지가 얼마나 많은 색상을 표현할 수 있는가?__ 를 의미한다.       
- 정밀도가 높을 수록 많은 색상을 표현할 수 있어 데이터 폭이 넓어지고, 더 자연스럽게 표시된다. 데이터의 처리 결과도 더 정밀해진다.   

- 비트 : 색상의 표현 개수를 설정     
- 표현 : nBit, nBit의 의미는 $2^n$ 이므로, 4bit는 $2^4$인, 16가지 색상 표현이 가능하다.      

> 색상을 표현할 때 적어도 몇 비트가 필요한가요?         

> n.uint8이 표현할 수 있는 범위를 정수로 표현하면 얼마인가요?
